```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
help(pd.merge)
```

```python
pd.merge(registrations, logins, how = 'inner', on = 'name')
# inner join only takes what is present in both tables; the column on which the inner join is done needs to have unique values
```

```python
# left and right merge
# the order of the tables matter here; left - everything that is present in the left hand table, if there is nor result for the
# ON column in the right table, it will fill in with NaN; 
# right - everything that is present in the right hand table, if there is no result for the
# ON column in the left table, it will fill in with NaN; 
# 
```

```python
pd.merge(left = registrations, right = logins, how = 'left', on = 'name')
```

```python
pd.merge(left = registrations, right = logins, how = 'right', on = 'name')
```

```python
# pandas is also able to identify the ON column
```

```python
# outer merge - allows to include everything that is present in both tabels
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations, logins, how = 'outer', on = 'name')
```
