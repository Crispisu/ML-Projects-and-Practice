```python
df_dummies = pd.get_dummies(df.drop('name', axis=1))
```

```python
df_dummies.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_dummies)
```

```python
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=df_dummies.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize=(15,8), dpi=150)
sns.heatmap(scaled_df)
```

```python
len(df)
```

```python
sns.clustermap(scaled_df, col_cluster=False)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=4)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
plt.figure(figsize=(12,4), dpi=150)
sns.scatterplot(data=df, x='mpg', y='weight', hue=cluster_labels)
```

```python
model = AgglomerativeClustering(n_clusters=None, distance_threshold=0)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

```python
plt.figure(figsize=(20,10), dpi=150)

dendro = dendrogram(linkage_matrix, truncate_mode='level', p=3)
```
