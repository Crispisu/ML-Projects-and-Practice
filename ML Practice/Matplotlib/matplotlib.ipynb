```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig, axes = plt.subplots(2, 2)

axes[0][0].plot(x, y)
axes[1][1].plot(a, b)
```

```python
type(axes)
```

```python
axes.shape
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 1)

for ax in axes:
    ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 1)

axes[0].plot(x,y)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y Label 1,0')
axes[1][0].set_xlim(2,6)
axes[1][1].plot(a,b)
axes[1][1].set_title('Title 1,1')

fig.suptitle('Figure Level', fontsize = 16)

fig.subplots_adjust(wspace = 0.5, hspace = 0.5)

fig.savefig('new_subplots.png', bbox_inches = 'tight')
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label = 'X vs X')

ax.plot(x,x**2, label = 'X vs X^2')
ax.legend(loc = (1.1, 0.5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color = '#e6c017', label = 'x v x') 
#  you can either put the color you want, or put in the Hex colour code for a customized colour

ax.plot(x, x+1, color = 'purple', label = 'x v x+1')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color = '#e6c017', lw = 10) 
#  you can either put the color you want, or put in the Hex colour code for a customized colour

ax.plot(x, x+1, color = 'purple', linewidth = 5)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color = '#e6c017', lw = 10, ls = '--') 
#  you can either put the color you want, or put in the Hex colour code for a customized colour

ax.plot(x, x+1, color = 'purple', linewidth = 5, linestyle = '-.')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# ax.plot(x,x, color = '#e6c017', lw = 10, ls = '--') 
#  you can either put the color you want, or put in the Hex colour code for a customized colour

lines = ax.plot(x, x+1, color = 'purple', linewidth = 5)
lines[0].set_dashes([5,2,10,2])
# 5 solid points then 2 blank points then 10 solid and 2 blank
```
